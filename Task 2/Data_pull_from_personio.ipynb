# **Pre-process employee json to csv format which can be used for further analysis**

```python
import pandas as pd
import json
data = json.loads(json_string_employees)
df_employee = pd.json_normalize(data['data'])

df_employee = df_employee[df_employee.columns.drop(list(df_employee.filter(regex='label')))]
```

```python
df_employee.columns
```

**Pre-process column information to readable form**

```python
df_employee.columns = df_employee.columns.str.replace(r'attributes.', '')
df_employee.columns = df_employee.columns.str.replace(r'.value', '')
df_employee.columns = df_employee.columns.str.replace(r'dynamic_1146702', 'Country')
df_employee.columns = df_employee.columns.str.replace(r'dynamic_1146666', 'Birthday')
```

```python
#pd.json_normalize(df_employee.absence_entitlement[0])

df_employee = pd.concat([df_employee, 
                pd.json_normalize(df_employee.absence_entitlement[0])], 
                axis=1).drop('absence_entitlement', 1)

```
